```python
# 
# Задание 3.1 
# Нарисуйте график функции k*cos(x-a)+b 3 различных значений параметров k, a, b
# 
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

# меняем размер графика
fig_size = plt.rcParams["figure.figsize"]
fig_size[0] = 16
fig_size[1] = 9

# задаем область рисования, от нуля до четырех пи, и количество точек
x = np.linspace(0,4*np.pi, 200)

# график с k=1, a=0, b=0
plt.plot(x, 1*np.cos(x-0)+0, label='cos(x), k*cos(x-a)+b, k=1, a=0, b=0 \n')

# график с k=2, a= -1/2pi, b=1/2
plt.plot(x, 2*np.cos(x+1/2*np.pi)+1/2, label='2*cos(x+1/2pi)+1/2, k*cos(x-a)+b, k=2, a= -1/2pi, b=1/2 \n')

# график с k=1/2, a=pi, b=-1/2
plt.plot(x, 1/2*np.cos(x-np.pi)-1/2, label='1/2*cos(x-pi)-1/2, k*cos(x-a)+b, k=1/2, a=pi, b=-1/2 \n')

# параметры визуализации графика
plt.xlabel('x')
plt.ylabel('y')
plt.legend(loc='upper left')
plt.axis()
plt.grid()
plt.ylim(-2,4)
plt.show()
```

```python
# 
# Задание 4.1 
# Решите систему уравнений
# y = x2 – 1
# exp(x) + x∙(1 – y) = 1
# 
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

# импортируем функцию для вычисления корней уравнений
from scipy.optimize import fsolve

# пишем функцию, возвращающую численное значение каждого уравнения в заданной точке
def equations(p):
    x, y = p
    return (y - x**2 + 1, np.exp(x) - x*y +x - 1)

# Корни системы уравнения 
# Подставляем функцию, которая возвращающую численное значение каждого уравнения в заданной точке
# в функцию, которая высчитывает корни уравнений
# и указываем отправные точки, выбирая их рядом с пересечениями графиков уравнений
x1, y1 = fsolve(equations, (-2, 2))
x2, y2 = fsolve(equations, (2, 4))
x3, y3 = fsolve(equations, (4, 15))

print ('Корни системы уравнения:')
print ('x1 = {:.5f}, y1 = {:.5f}'.format(x1, y1))
print ('x2 = {:.5f}, y2 = {:.5f}'.format(x2, y2))
print ('x3 = {:.5f}, y3 = {:.5f}'.format(x3, y3))
# построение графика 

# меняем размер графика
fig_size = plt.rcParams["figure.figsize"]
fig_size[0] = 16
fig_size[1] = 9

# задаем область рисования и количество точек
x = np.linspace(-2, 5, 201)

# график 1-го уравнения y=x^2-1
plt.plot(x, x**2-1, label='y=x^2-1')

# график 2-го уравнения y=(exp(x)+x-1)/x
plt.plot(x, (np.exp(x)+x-1)/x, label='y=(exp(x)+x-1)/x')

# визуализация корней уравнений
plt.plot(x1, y1, 'b*', label='x1, y1')
plt.plot(x2, y2, 'r*',label='x2, y2')
plt.plot(x3, y3, 'g*',label='x3, y3')

# параметры визуализации графика
plt.xlabel('x')
plt.ylabel('y')
plt.legend(loc='upper left')
plt.axis()
plt.grid()
plt.ylim(-4,20)
plt.show()
```
